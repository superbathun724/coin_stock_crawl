```python
import requests
import yfinance as yf
import pandas as pd

url = "https://api.coingecko.com/api/v3/simple/price"
params = {
    "ids": "bitcoin,ethereum,dogecoin",
    "vs_currencies": "usd,krw",
    "include_24hr_change": "true"
}
crypto_data = requests.get(url, params=params).json()
print("ðŸ’° Crypto Prices:")
print(crypto_data)

ticker = yf.Ticker("AAPL")
hist = ticker.history(period="7d", interval="1d")
print("\nðŸ“ˆ AAPL Stock Data (last 7 days):")
display(hist)

pd.DataFrame(crypto_data).to_csv("../data/crypto_data.csv")
hist.to_csv("../data/stock_data.csv")
print("\nâœ… Data saved to '../data/'")
```
